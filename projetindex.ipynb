```python
f1 = open("/content/drive/MyDrive/projet indexation/corpus/fichier1.txt", "r", encoding='utf-8-sig')
f2 = open("/content/drive/MyDrive/projet indexation/corpus/f2.txt", "r", encoding='utf-8-sig')
f3 = open("/content/drive/MyDrive/projet indexation/corpus/f3.txt", "r", encoding='utf-8-sig')
f4 = open("/content/drive/MyDrive/projet indexation/corpus/f4.txt", "r", encoding='utf-8-sig')
f5 = open("/content/drive/MyDrive/projet indexation/corpus/f5.txt", "r", encoding='utf-8-sig')
f6 = open("/content/drive/MyDrive/projet indexation/corpus/f6.txt", "r", encoding='utf-8-sig')
f7 = open("/content/drive/MyDrive/projet indexation/corpus/f7.txt", "r", encoding='utf-8-sig')
from nltk.tokenize import word_tokenize
mot1=word_tokenize(f1.read())
print("La liste des mots de doc1",mot1)
mot2=word_tokenize(f2.read())
print("La liste des mots de doc2",mot2)
mot3=word_tokenize(f3.read())
print("La liste des mots de doc3",mot3)
mot4=word_tokenize(f4.read())
print("La liste des mots de doc4",mot4)
mot5=word_tokenize(f5.read())
print("La liste des mots de doc5",mot5)
mot6=word_tokenize(f6.read())
print("La liste des mots de doc6",mot6)
mot7=word_tokenize(f7.read())
print("La liste des mots de doc7",mot7)
```

** Identification des mots vides**

```python
from nltk.corpus import stopwords
```

```python
vide1=set(stopwords.words('english'))
print(vide1)
```

```python
filtered_sentence1 = [w for w in mot1 if not w in vide1]
filtered_sentence2 = [w for w in mot2 if not w in vide1]
filtered_sentence3 = [w for w in mot3 if not w in vide1]
filtered_sentence4 = [w for w in mot4 if not w in vide1]
filtered_sentence5 = [w for w in mot5 if not w in vide1]
filtered_sentence6 = [w for w in mot6 if not w in vide1]
filtered_sentence7 = [w for w in mot7 if not w in vide1]
print(filtered_sentence1)
print(filtered_sentence2)
print(filtered_sentence3)
print(filtered_sentence4)
print(filtered_sentence5)
print(filtered_sentence6)
print(filtered_sentence7)
```

** Racinisation**

```python
from nltk.stem import PorterStemmer
Racin1=[]
Racin2=[]
Racin3=[]
Racin4=[]
Racin5=[]
Racin6=[]
Racin7=[]
ps=PorterStemmer()
for w in filtered_sentence1:
  Racin1.append(ps.stem(w))
for w in filtered_sentence2:
  Racin2.append(ps.stem(w))
for w in filtered_sentence3:
  Racin3.append(ps.stem(w))
for w in filtered_sentence4:
  Racin4.append(ps.stem(w))
for w in filtered_sentence5:
  Racin5.append(ps.stem(w))
for w in filtered_sentence6:
  Racin6.append(ps.stem(w))
for w in filtered_sentence7:
  Racin7.append(ps.stem(w))
Racin=[Racin1,Racin2,Racin3,Racin4,Racin5,Racin6,Racin7]
```
